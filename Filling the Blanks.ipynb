```python
import pandas as pd
import numpy as py
```

```python
df=pd.read_csv('PD Fill the Blanks challenge.csv')
```

```python
pd.options.display.max_rows=None
df
```

```python
df.shape
```

```python
# fill the blanks
df.loc[:,['Lesson Name','Subject']]=df.groupby(by=['Weekday','Teacher']).fillna(method='ffill').fillna(method='bfill').loc[:,['Lesson Name','Subject']]
```

```python
# calculate average
df_temp=df.groupby(by=['Weekday','Teacher']).mean()
df_temp.head()
```

```python
for i in range(len(df_temp.index)):
    df.loc[(df['Weekday']==df_temp.index[i][0])&(df['Teacher']==df_temp.index[i][1]),'Avg. Attendance per Subject & Lesson']=df_temp.loc[df_temp.index[i],'Attendance']
```

```python
# output 
df.to_csv('Output_fillingBlanks_byJupyter.csv', index=False)
```
